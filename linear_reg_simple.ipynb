```python
#linear regression:
#y=mx+c
#y-dependent variable
#x-independent variable
#m-slope(gradient,coefficent)
#c-constant(Y intecept)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
from sklearn.datasets import load_boston
df=load_boston()
```

```python
df
```

```python
x=pd.DataFrame(df.data,columns=df.feature_names)
```

```python
x
```

```python
#we need to select only 2 columns into seperate dtaframe as we are doing linear regression.
z=pd.DataFrame(x,columns=['RAD','TAX'])
```

```python
z
```

```python
%matplotlib inline
plt.xlabel('radius')
plt.ylabel('tax')
plt.scatter(z.RAD,z.TAX,color='red')
```

```python
reg=linear_model.LinearRegression()
#fit-we are giving the data values and giving in 2 brackets becoz we need to pass 2-d array so to reshape it we need to take the 
#--attributes as the array.
m=np.array(z['RAD']).reshape(-1,1)
n=np.array(z['TAX']).reshape(-1,1)
reg.fit(m,n)
```

```python
m.shape
```

```python
n.shape
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict(m)
```

```python
#y=mx+c
17.61831445*3+239.99269677
```

```python
p=reg.predict(m)
```

```python
p
```

```python
z['predict']=p
```

```python
z
```

```python
z.to_csv("output1.csv",index=False)
```

```python
%matplotlib inline
plt.xlabel('radius')
plt.ylabel('tax')
plt.scatter(z.RAD,z.TAX,color='red')
plt.plot(z.RAD,p)
```
